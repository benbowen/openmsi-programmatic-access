## Peak identification

### Similarities and differeneces with other peak identification problems often encountered in MS
Peak determination in MSI is a very different beast than for LC-MS based approaches.  xcms identifies peaks _in the time domain_ by looking for single $m/z$ channels that have "Gaussian" shapes in the retention time dimension.  This is inappropriate for MSI where there is no time dimension.

In certain cases these time-dimension based techniques could be extended to $x$ and $y$ dimensions, but only if peaks are expected to occur in space.  In most imaging applications, this will not happen: "Gaussian" shaped "mountains" of intensity at a certain $m/z$ values are the exception, not the rule.  Instead the intensities of $m/z$ channels will vary arbitrarily in accordance with the biological peculiarities of the sample being examined.  One case where Gaussian or pseudo-guassian peaks _may_ occur in space is in "spotting" applications like Trent is working on.  Even then, the required assumption is that "spots" in an image will be many pixels large.

The peak identification problem this notebook focuses on is identifying and centroiding peaks in a single-dimensional vector, presumed to consist of "profile"-mode MS data from any arbitrary pixel in an image.
